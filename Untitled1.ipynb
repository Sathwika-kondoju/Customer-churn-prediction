```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

sns.set(style="whitegrid")
```

```python
df = pd.read_excel("/content/churn_100_rows (2).xlsx")
df.head()
```

```python
print(df.info())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df = df.dropna()
```

```python
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title("Tenure vs Churn")
plt.show()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)[:,1]
```

```python
print("Logistic Regression Results")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_lr))
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
y_prob_dt = dt.predict_proba(X_test)[:,1]
```

```python
print("\nDecision Tree Results")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_dt))
```

```python
importance = pd.Series(dt.feature_importances_, index=X.columns)
importance.nlargest(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)[:,1]

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

# Accuracy
acc_lr = accuracy_score(y_test, y_pred_lr)
acc_rf = accuracy_score(y_test, y_pred_rf)

# ROC-AUC
roc_lr = roc_auc_score(y_test, y_prob_lr)
roc_rf = roc_auc_score(y_test, y_prob_rf)

print("Logistic Accuracy:", acc_lr)
print("Random Forest Accuracy:", acc_rf)

print("Logistic ROC-AUC:", roc_lr)
print("Random Forest ROC-AUC:", roc_rf)
```

```python
import matplotlib.pyplot as plt
```

```python
models = ['Logistic Regression', 'Random Forest']
accuracy = [acc_lr, acc_rf]
roc_auc = [roc_lr, roc_rf]
```

```python
x = range(len(models))
```

```python
plt.figure()
plt.bar(x, accuracy)
plt.xticks(x, models)
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure()
plt.bar(x, roc_auc)
plt.xticks(x, models)
plt.title("ROC-AUC Comparison")
plt.xlabel("Models")
plt.ylabel("ROC-AUC Score")
plt.show()
```

```python
import numpy as np

x = np.arange(len(models))
width = 0.35

plt.figure()

plt.bar(x - width/2, accuracy, width, label='Accuracy')
plt.bar(x + width/2, roc_auc, width, label='ROC-AUC')

plt.xticks(x, models)
plt.title("Model Comparison (Accuracy vs ROC-AUC)")
plt.xlabel("Models")
plt.ylabel("Score")
plt.legend()

plt.show()
```
